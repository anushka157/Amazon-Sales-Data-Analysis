```python
# Assuming df is your DataFrame and 'Size' is the column you're plotting
# Assign 'Size' to both x and hue, and set legend=False

ax=sns.countplot(x='Size',hue='Size',data=df , palette='viridis',legend=False)
# Add title and labels
plt.title('Count Plot with Custom Colors')
plt.xlabel('Size')
plt.ylabel('Count')
plt.show()
```

### Group By
#### the group by() function in pandas is used to group data based on one or more columns in a dataframe

```python
df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
```

```python
S_Qty=df.groupby(['Size'], as_index=False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sns.barplot(x='Size',y='Qty',data=S_Qty,hue='Size',palette='viridis',legend=False)
```

#### note: from above graph you can see that most of the qty buys M size in the sales

```python
#checking b2b data using pie chart
B2B_Check = df['B2B'].value_counts()
# plot the pie chart
plt.pie(B2B_Check , labels = B2B_Check, autopct = '%1.1f%%')

#plt.axix('equal')
plt.show()
```

```python
#checking b2b data using pie chart
B2B_Check = df['B2B'].value_counts()
# plot the pie chart
plt.pie(B2B_Check , labels = B2B_Check.index, autopct = '%1.1f%%')

#plt.axix('equal')
plt.show()
```

#### From the above chart we can see that most of the buyers are retailers and less no. of b2b

```python
# prepare data for scatter plot
x_data = df['Category']
y_data = df['Size']

#plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
# plot the count of cities by state
plt.figure(figsize=(18,6))
sns.countplot(data=df, x = 'ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
#top 10 states
top_10_state = df['ship-state'].value_counts().head(10)
# plot count of cities by state
#palette = sns.color_palette("tab20", len(top_10_state))
plt.figure(figsize=(18,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_state.index)], x = 'ship-state', hue='ship-state', palette='tab20', legend=False)
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

#### from the above graph you can see that the most of the buyers are from maharastra
